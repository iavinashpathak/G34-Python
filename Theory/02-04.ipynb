```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = load_iris(return_X_y=True)
```

```python
X = data[0]
y = data[-1]
cols = load_iris().feature_names
```

```python
df = pd.DataFrame(X, columns=cols)
df['target'] = y
```

```python
df.head()
```

```python
df.sort_values('sepal length (cm)', ascending=True)
```

```python
df.sort_values('sepal length (cm)', ascending=False)
```

```python
df.sort_values(by=['sepal length (cm)', 'petal length (cm)'], ascending=True)
```

```python
df.sort_values(by=['sepal length (cm)', 'petal length (cm)'], ascending=[False, True])
```
