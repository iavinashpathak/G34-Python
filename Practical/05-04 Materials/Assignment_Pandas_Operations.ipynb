# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
automobile_df['engine-location'].unique()
```

```python
automobile_df[['engine-location', 'price']].loc[0]
```

```python
# Your code here. Use apply and lambda
def update_price(value_pair):
    if value_pair['engine-location'] == 'rear':
        return value_pair['price']*2
    else:
        return value_pair['price']

automobile_df['price'] = automobile_df[['engine-location', 'price']].apply(lambda val: update_price(val), axis=1)
```

```python
print(f'Before Updation, Price of car with rear engines: {df_copy[df_copy['engine-location'] == 'rear']['price'].values}\nAfter Updation, Price of car with rear engines{automobile_df[automobile_df['engine-location'] == 'rear']['price'].values}')
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>

<code>GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}</code>

<code>japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}</code>

```python
# Your code here
GermanCars = pd.DataFrame({'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]})
japaneseCars = pd.DataFrame({'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]})
```

```python
#row-wise
pd.concat([GermanCars, japaneseCars], axis=0)
```

```python
#column-wise
pd.concat([GermanCars, japaneseCars], axis=1)
```
